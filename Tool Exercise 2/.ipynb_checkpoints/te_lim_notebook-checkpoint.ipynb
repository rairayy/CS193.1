## Preprocessing Stage

### Cleaning

```python
# previously defined function for cleaning tweets from previous class notebooks

def clean_tweets(tweets):
    """
    Replaces empty tweets, replaces text with lower case characters,
    remove special characters and RTs, remove leading and trailing
    whitespaces, and remove stopwords.
    """
    tweets['cleaned_text'] = tweets['text'].fillna('')
    tweets['cleaned_text'] = tweets['cleaned_text'].str.lower()
    tweets['cleaned_text'] = tweets['cleaned_text'].str.replace(r'(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|rt', '')
    tweets['cleaned_text'] = tweets['cleaned_text'].str.replace(r'^\s+|\s+$', '') 
    tweets['cleaned_text'] = tweets['cleaned_text'].apply(lambda x: ' '.join([w for w in x.split() if w not in (stopwords)]))
    return tweets
```

```python
cleaned_tweets = clean_tweets(tweets)
cleaned_tweets.head()
```

### Spelling Correction

```python
def correct():
    # last df is cleaned_tweets
    corrected = cleaned_tweets
    corrected['corrected'] = ''
    for i in range(0,len(corrected.index)):
        corrected['corrected'][i] = str(tb(corrected['cleaned_text'][i]).correct())
    return corrected
```

```python
corrected_tweets = correct()
corrected_tweets.head()
```

### Text Lemmatization

```python
# Function for lemmatizing
def lemmatize():
    # last df is corrected_tweets
    tweets = corrected_tweets
    tweets['lemmatized'] = ''
    for i in range(0,len(tweets.index)):
        lemmatized = ''
        blob = tb(tweets['corrected'][i]).sentences[0]
        tags = blob.tags
        for j in range(0,len(tags)):
            if tags[j][1].startswith('NN'):
                lemmatized += tags[j][0].lemmatize('n')
                lemmatized += ' '
            elif tags[j][1].startswith('VB'):
                lemmatized += tags[j][0].lemmatize('v')
                lemmatized += ' '
            elif tags[j][1].startswith('JJ'):
                lemmatized += tags[j][0].lemmatize('a')
                lemmatized += ' '
            else:
                lemmatized += tags[j][0]
                lemmatized += ' '
        tweets['lemmatized'][i] = lemmatized.strip()
    return tweets
```

```python
lemmatized = lemmatize()
lemmatized.head()
```

## Vectorization

```python
def text_to_cv(tweets):
    vectorizer = CountVectorizer(max_features=10000)
    tfidf = vectorizer.fit_transform(tweets)
    return tfidf
```

```python
tfidf = text_to_cv(lemmatized['lemmatized'])
print('TFIDF matrix shape:', tfidf.shape)
```

## Classification

Using only the cleaned text a feature

```python
# splitting data
text_train, text_test, label_train, label_test = train_test_split(tfidf, lemmatized['airline_sentiment'], test_size=0.2)
text_train.shape, label_train.shape
```

### Logistic Regression

```python
# training
log_model = LogisticRegression().fit(text_train, label_train)
```

```python
# predicting
log_prediction = log_model.predict(text_test)
log_prediction
```

```python
# evaluating
print(classification_report(label_test, log_prediction))
```

### Random Forest Classifier

```python
# training
rf_model = RandomForestClassifier().fit(text_train, label_train)
```

```python
# predicting
rf_prediction = rf_model.predict(text_test)
```

```python
# Evaluation
print(classification_report(label_test,rf_prediction))
```

### Support Vector Classifier

```python
# training
svc_model = SVC(kernel="linear", C=0.025).fit(text_train, label_train)
```

```python
# predicting
svc_pred = svc_model.predict(text_test)
```

```python
# evaluation
print(classification_report(label_test,svc_pred))
```
