```python
import json
import networkx as nx
from networkx.readwrite import json_graph
```

```python
def read_json_file(filename):
    with open(filename) as f:
        graph = json.load(f)
    return json_graph.node_link_graph(graph)
```

```python
G = read_json_file('hp.json')
```

```python
import matplotlib.pyplot as plt
pos = nx.spring_layout(G,k=0.5,iterations=10)
#pos = nx.spring_layout(G,scale=2)
plt.figure(3,figsize=(12,12))
nx.draw_networkx(G,pos,node_size=160,font_size=8,width=0.2)
```

```python
nodeDict = dict(G.nodes(data=True))
for x in nodeDict:
    print(x+'   '+nodeDict[str(x)]['label'])
```

## Main Trio Analysis

#### Clustering

```python
graph = nx.Graph(G)
print('Local Harry:', nx.clustering(graph, '106'))
print('Local Hermoine:', nx.clustering(graph, '59'))
print('Loacl Ron:', nx.clustering(graph, '147'))
print('Local Victor:', nx.clustering(graph, '71'))
```

```python
print('Global:', nx.average_clustering(graph))
print('Transitivity:', nx.transitivity(graph))
```

#### General Graph Stuff

```python
# Average distance
print('Average distance:', nx.average_shortest_path_length(graph))

# Diameter
print('Diameter:', nx.diameter(graph))

# Eccentricity
print('Eccentricity:', nx.eccentricity(graph))

# Radius
print('Radius:', nx.radius(graph))

# Periphery
print('Periphery nodes:', nx.periphery(graph))

# Center
print('Center nodes:', nx.center(graph))
```

#### Degree Centrailty

```python
dc = nx.degree_centrality(graph)
print('Harry: ', dc['106'])
print('Hermoine: ', dc['59'])
print('Ron: ', dc['147'])
```

#### Closeness Centrality

```python
cc = nx.closeness_centrality(graph)
print('Harry: ', cc['106'])
print('Hermoine: ', cc['59'])
print('Ron: ', cc['147'])
```
