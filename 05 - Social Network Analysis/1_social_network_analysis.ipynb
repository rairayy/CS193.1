### Network configuration

```python
# Undirected Graph
graph = nx.Graph()
graph.add_edge('A', 'B')
graph.add_edge('B', 'C')
nx.draw(graph, with_labels=True)
```

```python
# Directed Graph (Digraph)
digraph = nx.DiGraph()
digraph.add_edge('A', 'B')
digraph.add_edge('B', 'C')
nx.draw(digraph, with_labels=True)
```

```python
# Weighted graph
weighted = nx.Graph()
weighted.add_node(1, pos=(1, 1))
weighted.add_node(2, pos=(2, 2))
weighted.add_node(3, pos=(1, 0))
weighted.add_edge(1, 2, weight=6)
weighted.add_edge(1, 3, weight=3)

pos = nx.get_node_attributes(weighted, 'pos')
nx.draw(weighted, pos, with_labels=True)
weights = nx.get_edge_attributes(weighted, 'weight')
nx.draw_networkx_edge_labels(weighted, pos, edge_labels=weights)
```

```python
other = nx.Graph()
other.add_node('A', pos=(0, 1))
other.add_node('B', pos=(0, 2))
other.add_node('C', pos=(2, 1))
other.add_node('D', pos=(1, 2))
other.add_node('E', pos=(2, 2))


other.add_edge('A', 'B', relation='friend')
other.add_edge('B', 'C', relation='neighbor')
other.add_edge('C', 'D', relation='coworker')
other.add_edge('D', 'E', relation='family')
other.add_edge('D', 'A', relation='spouse')

pos = nx.get_node_attributes(other, 'pos')
nx.draw(other, pos, with_labels=True)
relations = nx.get_edge_attributes(other, 'relation')
nx.draw_networkx_edge_labels(other, pos, edge_labels=relations)
```

### Network attributes

##### Edge attributes

```python
G = nx.Graph()
G.add_edge('A', 'B', weight=6, relation='family')
G.add_edge('B', 'C', weight=3, relation='friend')

# Get all edges
print(G.edges())
# Get all edges with all attribute values
print(G.edges(data=True))
# Get all edges with only specific attributes values
print(G.edges(data='relation'))

# Get attributes of specific edge
print(G.get_edge_data('A', 'B'))
# Get specific attribute value of an edge
print(G.get_edge_data('A', 'B')['relation'])
```

##### Node attributes

```python
G = nx.Graph()
G.add_edge('A', 'B', weight=6, relation='friend')
G.add_edge('B', 'C', weight=6, relation='co-worker')

G.add_node('A', role='manager')
G.add_node('B', role='trader')
G.add_node('C', role='analyst')

# Access node data in NetworkX
print(G.nodes())
print(G.nodes(data=True))
print(G.nodes(data='role'))
```

### Clustering coefficient

```python
G = nx.Graph()
G.add_edges_from([('A', 'K'), ('A', 'B'), ('A', 'C'), ('B', 'C'),
                  ('B', 'K'), ('C', 'E'), ('C', 'F'), ('D', 'E'),
                  ('E', 'F'), ('E', 'H'), ('F', 'G'), ('I', 'J')])
nx.draw(G, with_labels=True)
```

#### Local Clustering Coefficient
- Proportion of connections among a node's neighbours which are actually realised compared with the number of all possible connections
- Higher clustering coefficient: more neighbors of the node are also neighbors
- Quantifies how close its neighbours are to being a clique (complete graph)

#### Global Clustering Coefficient
- Based on triplets of nodes

![global clustering](https://wikimedia.org/api/rest_v1/media/math/render/svg/7a5e18001f776670ede7612bbf6db5b320f3a153)

```python
'''
Higher clustering coefficient: more neighbors of the node are also neighbors
"quantifies how close its neighbours are to being a clique (complete graph)"
'''
# Local clustering coefficient
print('Local A:', nx.clustering(G, 'A'))
print('Local F:', nx.clustering(G, 'F'))
print('Loacl J:', nx.clustering(G, 'J'))

'''
based on triplets of nodes
'''
# Global clustering coefficient
print('Global:', nx.average_clustering(G))

# Transitivity
print('Transitivity:', nx.transitivity(G))
```

### Distance

```python
G = nx.Graph()
G.add_edges_from([('A', 'K'), ('A', 'B'), ('A', 'C'), ('B', 'C'),
                  ('B', 'K'), ('C', 'E'), ('C', 'F'), ('D', 'E'),
                  ('E', 'F'), ('E', 'H'), ('F', 'I'), ('I', 'J')])
nx.draw(G, with_labels=True)
```

#### Graph Eccentricity
- Maximum graph distance between `v` and any other vertex `u` of `G`

#### Diameter
- Maximum eccentricity in the graph `G`

#### Radius
- The radius of a graph is the minimum graph eccentricity of any graph vertex in a graph. A disconnected graph therefore has infinite radius

#### Periphery
- Periphery of a graph `G` is the subgraph of `G` induced by vertices that have graph eccentricities equal to the graph diameter.

#### Center
- The center of a graph `G` is the set of vertices of graph eccentricity equal to the graph radius (i.e., the set of central points).

```python
# Distance
print('Path from A to E:', nx.shortest_path(G, 'A', 'E'))
print('Distance from A to E:', nx.shortest_path_length(G, 'A', 'E'))

# Average distance
print('Average distance:', nx.average_shortest_path_length(G))

# Diameter
print('Diameter:', nx.diameter(G))

# Eccentricity
print('Eccentricity:', nx.eccentricity(G))

# Radius
print('Radius:', nx.radius(G))

# Periphery
print('Periphery nodes:', nx.periphery(G))

# Center
print('Center nodes:', nx.center(G))
```

### Centrality measure

```python
undirected = nx.karate_club_graph()
undirected = nx.convert_node_labels_to_integers(undirected, first_label=1)
nx.draw(undirected, with_labels=True)
```

#### Centrality
- identify the most important vertices within a graph
- most influential person(s), key nodes, etc

#### Degree Centrality
- Historically first and conceptually simplest is degree centrality, which is defined as the **number of links incident upon a node** (i.e., the number of ties that a node has)
- The degree can be interpreted in terms of the immediate risk of a node for catching whatever is flowing through the network (such as a virus, or some information)

```python
# Degree centrality for undirected graph
dc = nx.degree_centrality(undirected)
print('DC 34:', dc[34])
print('DC 33:', dc[33])
print('DC 17:', dc[17])
```

```python
directed = nx.DiGraph()
directed.add_edges_from([('A', 'K'), ('A', 'B'), ('A', 'C'), ('B', 'C'),
                  ('B', 'K'), ('C', 'E'), ('C', 'F'), ('D', 'E'),
                  ('E', 'F'), ('E', 'H'), ('F', 'G'), ('I', 'J')])
nx.draw(directed, with_labels=True)
```

#### Degree Centrality Cont'd
- In the case of a directed network (where ties have direction), we usually define two separate measures of degree centrality, namely **indegree** and **outdegree**.
- **Indegree** - a count of the number of ties directed to the node
- **Outdegree** - the number of ties that the node directs to others.

```python
# Degree centrality for digraph
in_dc = nx.in_degree_centrality(directed)
print(in_dc)
out_dc = nx.out_degree_centrality(directed)
print(out_dc)
```

#### Closeness Centrality
- The average length of the shortest path between the node and all other nodes in the graph
- The more central a node is, the closer it is to all other nodes
- How easily can the other node reach a particular node

![closeness centrality](https://wikimedia.org/api/rest_v1/media/math/render/svg/c01118e92aaf698613cf58c3d9062b029c56759f)

Where `d(y,x)` is the distance between vertices `x` and `y`. This formula is multiplied by `n-1` where `n` is the number of nodes.

```python
# Closeness centrality
nx.closeness_centrality(directed)
```

#### Betweenness Centrality
- A centrality measure of a vertex within a graph
- Quantifies the number of times a node acts as a bridge along the shortest path between two other nodes
- Computed as follows:
    1. For each pair of vertices `(s,t)`, compute the shortest paths between them.
    2. For each pair of vertices `(s,t)`, determine the fraction of shortest paths that pass through the vertex in question (here, vertex `v`).
    3. Sum this fraction over all pairs of vertices `(s,t)`.

```python
nx.draw(directed, with_labels=True)
```

```python
# Betweenness centrality
nx.betweenness_centrality(directed)
```
